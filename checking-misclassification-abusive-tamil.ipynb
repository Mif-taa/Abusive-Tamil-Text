```python
import pandas as pd

# Load true labels and predictions
test_df = pd.read_csv('/kaggle/input/datasets/rishta19/testing/test_with_labels.csv')  # has 'text' and 'label' columns
pred_df = pd.read_csv('/kaggle/input/datasets/rishta19/testing/CUET_Synthetica_FINAL_ENSEMBLE_12models.csv')  # has 'predicted_label' column

# Combine
test_df['predicted'] = pred_df['Class']

# Find misclassified
misclassified = test_df[test_df['Class'] != test_df['predicted']]

# Split into false negatives (missed abuse) and false positives (over-flagged)
false_negatives = misclassified[
    (test_df['Class'] == 'Abusive') & 
    (test_df['predicted'] == 'Non-Abusive')
]

false_positives = misclassified[
    (test_df['Class'] == 'Non-Abusive') & 
    (test_df['predicted'] == 'Abusive')
]

print(false_negatives[['Text', 'Class', 'predicted']].head(10))
```
